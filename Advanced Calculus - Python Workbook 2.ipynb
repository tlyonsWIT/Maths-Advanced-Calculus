# Surface plots of 2-variable functions

* A 2 varaible function $f(x,y)$ is one which depends on two-variables $x$ and $y$.


* The $(x,y)$ variables are usually plotted in the horizontal plane. 


* The __value__ of the function $f(x,y)$ representes the __height__ of the surface above the point $(x,y).$

## Creating the $(x,y)$ plane

* To create an $x$-axis we use the __numpy__ function __linspace__ 

```python
X=np.linspace(-10,10,251)
```

* This creates a list of numbers, starting at -10, and taking 251 steps to get to +10 (see code below)

```python
X
```

* We create a $y$-axis in the same way:

```python
Y=np.linspace(-10,10,251)
```

### The $(x,y)$-plane


* We combine these axes into a collection of $(x,y)$-pairs using the __numpy__ function __meshgrid.__

```python
x,y=np.meshgrid(X,Y)
```

* There are $251\times251=63001$ individual __(x, y)__ points formed by this meshgrid  

## Defining and Plotting a function


* We are now in a position to define a 2-variable function $f(x,y)$ in Python.


### Example 1: 

__(1)__ Define a Python function to represent the 2-variable function

$$f(x,y)=x^2-y^2.$$

__(2)__ Create a surface plot of this function.


```python
def f(x,y):
    return x**2-y**2
```

```python
fig=plt.figure(figsize=(9,6)) # Create a figure environment with a specified size
ax=fig.gca(projection='3d') # Instruct Python to render this figure in 3D
ax.plot_surface(x,y,f(x,y),color='cornflowerblue',alpha=0.5) # Plot this function f(x,y) over the (x,y) meshgrid
ax.set_xlabel('x',size=16) # Set the x-label
ax.set_ylabel('y',size=16) # Set the y-label
ax.set_zlabel('z',size=16) # Set the z-label
```

* In the __ax.plot_surface__ command the __color__ represents the colour of the surface.

* The __alpha__ paramter sets the transparency:
    
    * 0 makes the surface invisible.
    
    * 1 makes the surface completely opaque.
    
    
### Alternative surface plot

* Instead of plotting the function in the form of a surface, we can plot the function using a __wireframe__ in a similar way.

```python
fig=plt.figure(figsize=(9,6)) # Create a figure environment with a specified size
ax=fig.gca(projection='3d') # Instruct Python to render this figure in 3D
ax.plot_wireframe(x,y,f(x,y),color='cornflowerblue',alpha=0.5) # Plot this function f(x,y) over the (x,y) meshgrid
ax.set_xlabel('x',size=16) # Set the x-label
ax.set_ylabel('y',size=16) # Set the y-label
ax.set_zlabel('z',size=16) # Set the z-label
```

# Section 2 - Contours and level-sets

* The __contours__ of a surface plot are the curves of __constant height__ along the surface.

* We can plot these contours using the __contour__ function.


## Example 2

* Plot the surface $f(x,y)=x^2-y^2$ and its contours.

```python
fig = plt.figure(figsize=(12,9))
ax = fig.gca(projection='3d')
ax.plot_surface(x,y,f(x,y),color='cornflowerblue',alpha=0.25)
ax.contour(x, y, f(x,y), cmap=cm.coolwarm)
ax.set_xlabel('x',size=16) # Set the x-label
ax.set_ylabel('y',size=16) # Set the y-label
ax.set_zlabel('z',size=16) # Set the z-label
```

* The option __cm.coolwarm__ is the __color map__ called bone __coolwarm__


* A list of color maps is available at  [__Python colormaps__](https://matplotlib.org/tutorials/colors/colormaps.html)


## Level-sets

* The __level-sets__ of this plot are the contours of the surface __projected__ to the $(x,y)$-plane.


* These can be plotted using the __contour__ function again.


* The difference this time is we use the extra options __zdir='z'__ and __offset=-40__ to say the contours should be plotted on the $(x,y)$-plane at height $z=-40$ (offset).




## Example 3

* Plot the function $f(x,y)=x^2-y^2$ with its contours __and__ its level-sets.

```python
fig = plt.figure(figsize=(12,9))
ax = fig.gca(projection='3d')
ax.plot_surface(x,y,f(x,y),color='cornflowerblue',alpha=0.25) # Plot the surface 

ax.contour(x, y, f(x,y), cmap=cm.bone) # Plot the contours on the surface

ax.contour(x, y, f(x,y), zdir='z', offset=-40, cmap=cm.copper) # Plot the level curves in the (x,y)-plane

ax.set_xlabel('x',size=16) # Set the x-label
ax.set_ylabel('y',size=16) # Set the y-label
ax.set_zlabel('z',size=16) # Set the z-label

ax.set_zlim(-40,20) # set the vertical limits of the plot at -40 (lower) and +20 (upper)
```

# Section 3 - Level-sets and gradient vectors

* The __gradient__ of the function $f(x,y)=x^2-y^2$ is given by

$$\nabla f=2x\hat{\imath}-2y\hat{\jmath}$$


* This gradient is an example of a __vector field,__ which means we have a vector associated with each point of the $(x,y)$-plane.


* The components of this vector are $2x$ along the $x$-axis and $-2y$ alog the $y$-axis.


* These vectros should __always__ be perpendicular to the corresponding points on the level sets of the function.


* We will plot these level sets as usual and the vector field using the __quiver__ function.


* The __quiver__ function has the following usage: __quiver(x,y,u,v)__
    * __x__ and __y__ are the coordinate of each vector in the vector field.
    * __u__ and __v__ are the horizontal and vertical components respectively,at each of these locations

## Example 4

* Plot the level sets of the function $f(x,y)$ in the $(x,y)$-plane.


* Plot the gradient field $\nabla f = 2x\hat{\imath}-2y\hat{\jmath}$ on the same plot.

```python
plt.figure(figsize=(12,9))

plt.contour(x, y, f(x,y)) # The level curves in the (x,y)-plane

plt.quiver(x[::10,::10],y[::10,::10],2*x[::10,::10],-2*y[::10,::10]) # The quiver plot for the gradient field

plt.xlabel('x',size=16) # The x label

plt.ylabel('y',size=16) # The y label
```

* The option __[::10,::10]__ instructs __quiver__ to only plot a vector at every 10th point along the $x$-axis and at every 10th point along the $y$-axis. 

* If we did not include this option then the vector plot would be too dense and it would be difficult to see with the vetor field or the contours.


#### Observation: It should be clear from the plot above that the gradient vector is always perpendicular to the level curves of the function.
