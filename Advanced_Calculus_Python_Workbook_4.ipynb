# Paramterised Paths and Vector Fields

## Calling the python libraries and functions


* In the cell below the necessary Python libraries and functions are imported, which will be needed for the current work book


```python
import matplotlib.pyplot as plt  ## Library of functions for 2D plotting
from matplotlib import cm as cm  ## Library of functions for ColourMaps 
from mpl_toolkits.mplot3d import Axes3D ## Library of functions for 3D plotting

import numpy as np ## Library of function for numerical work
from numpy import abs, sin, cos, log, exp, tan, cosh, sinh, tanh,sqrt ## Useful numerical function called from the NUMPY library
```

## Creating the time-parameter

```python
t=np.linspace(0,10,101)
```




## Creating the paramterised path

* We are going to plot a vector field along the parameterised path

$$\vec{r}(t)=\sqrt{t}\hat{\imath}+(3t+1)\hat{\jmath}+\frac{t}{2}\hat{k}\quad\text{  for  }\quad 0\leq t \leq 10. $$


* We represent the path using __3__ distinct functions, one for each component.

  

```python
def r1(t):
    return sqrt(t)

def r2(t):
    return 3*t+1

def r3(t):
    return t/2
```

## Creating the vector field

* The vector field we want to plot along this trajectory is given by

$$\vec{F}(x,y,z)=\sqrt{xy}\hat{\imath}-\sqrt{yz}\hat{\jmath}+\sqrt{xz}\hat{k}.$$



* To represent this vector field, we again create __3__ distinct functions, one for each component of the vector field.

```python
def F1(x,y,z):
  return sqrt(x*y)

def F2(x,y,z):
  return -sqrt(y*z)

def F3(x,y,z):
  return sqrt(x*z)
```

```python
n=10

fig=plt.figure(figsize=(9,6)) # Create a figure environment with a specified size
ax=fig.gca(projection='3d') # Instruct Python to render this figure in 3D

ax.plot(r1(t),r2(t),r3(t),color='black') ## This plots the curve

ax.quiver(r1(t[::n]),r2(t[::n]),r3(t[::n]),F1(r1(t[::n]),r2(t[::n]),r3(t[::n])),F2(r1(t[::n]),r2(t[::n]),r3(t[::n])),F3(r1(t[::n]),r2(t[::n]),r3(t[::n])),length=0.1,arrow_length_ratio=0.1,color='red') 
# This plots the vector field along the curve
ax.set_xlabel('x',size=16) # Set the x-label
ax.set_ylabel('y',size=16) # Set the y-label
ax.set_zlabel('z',size=16) # Set the z-label


#ax.set_xlim(0,400)
#ax.set_ylim(0,70)
#ax.set_zlim(-5,70)
```

## Plotting the velocity vector

* We can also plot the velocity vector at each of these points. 

* We see that the derifavtive of the vector $\vec{r}(t)$ is

$$\frac{d\vec{r}}{dt}=\frac{1}{2\sqrt{t}}\hat{\imath}+3\hat{\jmath}+\frac{1}{2}\hat{k}$$


* Again, we create __3__ distinct function to represent this velocity vector as follows

```python
def dr1(t):
  return 1/(2*sqrt(t))

def dr2(t):
  return 3


def dr3(t):
  return 1/2
```

* We can now plot an additional quiver plot on the previous plot, to represent this velocity vector. 



```python
n=10

fig=plt.figure(figsize=(9,6)) # Create a figure environment with a specified size
ax=fig.gca(projection='3d') # Instruct Python to render this figure in 3D

ax.plot(r1(t),r2(t),r3(t),color='black') ## This plots the curve

ax.quiver(r1(t[::n]),r2(t[::n]),r3(t[::n]),F1(r1(t[::n]),r2(t[::n]),r3(t[::n])),F2(r1(t[::n]),r2(t[::n]),r3(t[::n])),F3(r1(t[::n]),r2(t[::n]),r3(t[::n])),length=0.1,arrow_length_ratio=0.1,color='red') 
# This plots the vector field along the curve

ax.quiver(r1(t[::n]),r2(t[::n]),r3(t[::n]),dr1(t[::n]),dr2(t[::n]),dr3(t[::n]),length=0.9,arrow_length_ratio=0.1,color='blue') 
# This plots the velocity vector along the curve
ax.set_xlabel('x',size=16) # Set the x-label
ax.set_ylabel('y',size=16) # Set the y-label
ax.set_zlabel('z',size=16) # Set the z-label

```
