## Creating the $(x,y)$ plane

* We create the $(x,y)$-plane using the functions __linspace__ and __meshgrid__ from the __numpy__ library (see Worbook 2).

```python
X=np.linspace(-5,5,51)
Y=np.linspace(-5,5,51)
x,y=np.meshgrid(X,Y)
```

# Level-curves and gradient vectors


## Example 1

* Sketch the vector field
$$
\vec{D}=x\hat{\imath} + y\hat{\jmath}
$$



```python
n=5
xn=x[::n,::n]
yn=y[::n,::n]

plt.figure(figsize=(12,12))

plt.quiver(xn,yn,xn,yn) # The quiver plot for the gradient field

plt.xlabel('x',size=16) # The x label

plt.ylabel('y',size=16) # The y label
```

# Example 2

Plot the vector field
$$
\vec{R}=y\hat{\imath}-x\hat{\jmath}
$$

```python
n=5
xn=x[::n,::n]
yn=y[::n,::n]

plt.figure(figsize=(12,12))

plt.quiver(xn,yn,yn,-xn) # The quiver plot for the gradient field

plt.xlabel('x',size=16) # The x label

plt.ylabel('y',size=16) # The y label
```

# Vecor curls


# Example 1

Sketch the vector field
$$
\vec{F}(x,y,z)=xy\hat{\imath} + yz\hat{\jmath} - xz\hat{k}
$$

```python
X=np.linspace(-1,1,11)
Y=np.linspace(-1,1,11)
Z=np.linspace(-1,1,11)
x,y,z=np.meshgrid(X,Y,Z)
```

```python
n=5
xn=x[::n]
yn=x[::n]
zn=z[::n]
fig = plt.figure(figsize=(12,9))
ax = fig.gca(projection='3d')
ax.quiver(xn,yn,zn,xn*yn,yn*zn,-xn*zn)
ax.set_xlim(-1.5,1.5)
ax.set_ylim(-1.5,1.5)
ax.set_zlim(-1.5,1.5)



ax.set_xlabel('x',size=16) # Set the x-label
ax.set_ylabel('y',size=16) # Set the y-label
ax.set_zlabel('z',size=16) # Set the z-label
```

# Example 2

Sketch the vector field
$$
\vec{H}(x,y,z)=x\hat{\imath} + y\hat{\jmath} + z\hat{k}
$$

```python
n=5
xn=x[::n]
yn=x[::n]
zn=z[::n]
fig = plt.figure(figsize=(12,9))
ax = fig.gca(projection='3d')
ax.quiver(xn,yn,zn,xn,yn,zn)
ax.set_xlim(-1.5,1.5)
ax.set_ylim(-1.5,1.5)
ax.set_zlim(-1.5,1.5)



ax.set_xlabel('x',size=16) # Set the x-label
ax.set_ylabel('y',size=16) # Set the y-label
ax.set_zlabel('z',size=16) # Set the z-label
```
