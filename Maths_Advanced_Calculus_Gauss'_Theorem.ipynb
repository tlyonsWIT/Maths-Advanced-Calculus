# Python Workbook for Gauss' Theorem
## A.K.A. THe Divergence Theorem


* In this workbook we investigate vector fields in 2D of the form

$$\mathbf{\underline u}(x,y)=u(x,y)\hat{\imath}+v(x,y)\hat{\jmath}.$$


* The vector fields will always be investigated over __closed__ domains $\Omega$, whose boundary we denote $\partial \Omega$.

### Example 1:

Plot the vector field

$$ u(x,y) = e^{x}\hat{\imath}+e^{y}\hat{\jmath} $$

over the domain

$$\Omega: 0\leq x\leq 2\quad 0\leq y \leq 2.$$

We set up our $x$ and $y$ axes and the $(x,y)$-plane in the ususal way

```
X=np.linspace(-1,3,21) # The x-axis is made slightly bigger than the domain we are considering
Y=np.linspace(-1,3,21) # The y-axis is made slightly bigger than the domain we are considering
x,y=np.meshgrid(X,Y)
```

We now define the components of the vector field as follows

```
def u(x,y):
  return exp(x)
```

```
def v(x,y):
  return exp(y)
```

We can now plot the vecrot field using the function __quiver()__

```
plt.figure(figsize=(12,12))
ax=plt.gca()
plt.quiver(x,y,u(x,y),v(x,y))
ax.add_patch(Rectangle((0,0),2,2,fill=False))
```

### Example 3

* __Problem Set 5 -- Exercise 3__

The vector field from this exercise is given by
$$
\mathbf{\underline{F}}(x,y)=\frac{x^3}{6}\hat{\imath}+\frac{y^3}{6}\hat{\jmath}.
$$


We want to plot the flux of this vector field through the disk
$$
0\leq x^2+y^2\leq 1.
$$

#### Defining the vector components

```
def u(x,y):
  return x**3/6
```

```
def v(x,y):
  return y**3/6
```

```
X=np.linspace(-1.1,1.1,10)
Y=np.linspace(-1.1,1.1,10)
x,y=np.meshgrid(X,Y)
```

* The plot of the vector field is 

```
plt.figure(figsize=(6,6))
ax=plt.gca()
plt.quiver(x,y,u(x,y),v(x,y)) # The plot of the vector field
ax.add_patch(plt.Circle((0,0),1, fill=False, lw=2, ls='--',color='steelblue')) # The plot of the circle with centre (0,0) and radius 1.
```

* In this case we clearly see there is a net outward flux through the boundary.

* This means we expect the flux integral to be positive.

### Example 4

* Plot the vector field
$$
\mathbf{\underline{F}}(x,y)=y\sqrt{x^2+1}\hat{\imath}+y\hat{\jmath}
$$
over the region
$$\Omega: 0\leq \frac{x^2}{25}+\frac{y^2}{16}\leq 1.$$


* In this case the horizontal radius of the ellipse is $\sqrt{25}=5$.
* The vertical radius of the ellipse is $\sqrt{16}=4$.

* These radii will be useful when we want to plot the ellipse.

First we create an $(x,y)$-plane as follows

```
X=np.linspace(-5.5,5.5,10)
Y=np.linspace(-4.5,4.5,10)
x,y=np.meshgrid(X,Y)
```

Now we define the componetns of the vector field:

```
def u(x,y):
  return 5*x*y
```

```
def v(x,y):
  return 4*y*x
```

Now we plot the vectorfield using __quiver()__ and the boundary using __Ellipse()__

```
plt.figure(figsize=(6,6))
ax=plt.gca()
plt.quiver(x,y,u(x,y),v(x,y)) # The plot of the vector field
```

* We now define the $x$ and $y$ coordinates along the boundary of the ellipse as follows

```
def x_ellipse(theta):
  return 5*cos(theta)
```

```
def y_ellipse(theta):
  return 4*sin(theta)
```

* Now we need to define a range of values for the angle __theta__, which goes between $0^\circ$ and $360^\circ$, ot in terms of radians $0$ rad and $2\pi$ rad.

```
theta=np.linspace(0,2*pi,251) # Nore points means a smppther plot of the ellipse 
```

```
plt.figure(figsize=(10,8))
ax=plt.gca()
plt.quiver(x,y,u(x,y),v(x,y)) # The plot of the vector field
plt.plot(x_ellipse(theta,),y_ellipse(theta),lw=2,color='steelblue',ls='--')
```

* Again, we can clearly see there is a net flow out of the ellipse, we we expect the flus integral to be positive
