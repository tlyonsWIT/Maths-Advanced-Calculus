## Interpretation of the Curl

```
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

```
L=np.linspace(-1,1,10)

x, y, z = np.meshgrid(L,L,L)
```

```
u = -y
v = x
w = 0.5

a=0
b=0
c=2
```

```
fig = plt.figure(figsize=(15,15))
ax = fig.gca(projection='3d')
ax.quiver(x, y, z, u, v, w, length=0.1,color='cornflowerblue')
ax.quiver(x[::2,::2,::2], y[::2,::2,::2], z[::2,::2,::2], a, b, c, length=0.1,color='darkslategray')

#ax.set_xlim(-1,1)
#ax.set_ylim(-1,1)
#ax.set_zlim(-1,1)

plt.show()
```
